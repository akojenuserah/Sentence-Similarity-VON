# English Article Scraping

```python
# Voice of Nigeria - English

def von_en(start, end):
    categories = ["sport", "world", "business", "africa", "nigeria", "politics", "health", "entertainment & tourism"]
    data = []


    for category in categories:
      for pageNo in range(start, end):
        url = f"https://von.gov.ng/category/{category}/page/{pageNo}/"
        html_text = requests.get(url, headers=headers).text
        soup =BeautifulSoup(html_text,'html.parser')
        articles =soup.find_all('h2', class_="title")
        for article in articles:
            title = article.a.text.strip()
            link = article.a["href"]
            _article_page = requests.get(link, headers=headers).text
            page_soup = BeautifulSoup(_article_page,'html.parser')
            image_link = page_soup.find('meta', property='og:image')['content']
            _article_div = page_soup.find('div', class_='entry-content clearfix single-post-content')
            article_text = '\n'.join(p.get_text(strip=True) for p in _article_div.find_all('p') if p.get_text(strip=True))
            published_time = page_soup.find('meta', property='og:article:published_time')['content']
            modified_time = page_soup.find('meta', property='og:article:modified_time')['content']
            data.append({'Category': category,'Published_time':published_time, 'Title': title, 'Full_Content': article_text, 'article url':link, 'Image_link': image_link, 'Modified_time': modified_time})
    return data
```

```python
von_df = pd.DataFrame(von_en(1,2))
print(von_df.shape)
von_df
```

```python
def von_yor(start, end):
    categories = ["sport", "world", "business", "africa", "nigeria", "politics", "health", "entertainment"]
    data = []

    headers = {'User-Agent': 'Mozilla/5.0'}  # Assuming headers is defined somewhere, if not define it like this

    for category in categories:
        for pageNo in range(start, end):
            url = f"https://yoruba.von.gov.ng/category/{category}/page/{pageNo}/"
            html_text = requests.get(url, headers=headers, verify=False).text
            soup = BeautifulSoup(html_text, 'html.parser')
            articles = soup.find_all('h2', class_="title")
            for article in articles:
                title = article.a.text.strip()
                link = article.a["href"]
                _article_page = requests.get(link, headers=headers, verify=False).text
                _page_soup = BeautifulSoup(_article_page, 'html.parser')

                # Check if the meta tag exists
                meta_image = _page_soup.find('meta', property='og:image')
                image_link = meta_image['content'] if meta_image else "No image available"

                _article_div = _page_soup.find('div', class_='entry-content clearfix single-post-content')

                # Ensure the article div is found
                if _article_div:
                    article_text = '\n'.join(p.get_text(strip=True) for p in _article_div.find_all('p') if p.get_text(strip=True))
                else:
                    article_text = "No content available"

                # Check if the time tag exists
                time_tag = _page_soup.find('time', class_='post-published updated')
                published_time = time_tag['datetime'] if time_tag else "No published time available"

                data.append({'Published_time': published_time, 'Title': title, 'Full_Content': article_text, 'Article URL': link, 'Image_link': image_link})

    return data
```

```python
#YORUBA SPORTS

import requests
from bs4 import BeautifulSoup
import pandas as pd

data = []  # Initialize the list to store article data

for page in range(1, 2):  # Assuming you want to start with just the first page
    url = f"https://yoruba.von.gov.ng/category/sports/page/{page}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all articles on the current page
    articles = soup.find_all("article", class_="type-post")

    for article in articles:
        # Extract the required information from each article
        post_link_element = article.find("a", class_="img-cont")
        if post_link_element:
            post_link = post_link_element['href']
            post_title = post_link_element.get('title', '')

            # Fetch the full article text
            response = requests.get(post_link)
            article_soup = BeautifulSoup(response.content, 'html.parser')
            target_div = article_soup.find('div', class_='entry-content clearfix single-post-content')
            full_text = target_div.get_text(strip=True) if target_div else 'Full text not available'


            post_link = article.find("a", class_="img-cont")['href']
            post_title = article.find("a", class_="img-cont")['title']

            post_image = article.find("a", class_="post-thumbnail open-lightbox")['href']

            post_date = article.find("span", class_="time")


            post_summary_element = article.find("div", class_="post-summary")
            post_summary = post_summary_element.get_text().strip() if post_summary_element else ''


            # Append the extracted information to the data list
            data.append({
                "Title": post_title,
                "Date": post_date,
                "Summary": post_summary,
                "Image URL": post_image,
                "Post Link": post_link,
                "Full Text": full_text
            })

# Convert the list of dictionaries to a DataFrame
df2 = pd.DataFrame(data)

# Display the DataFrame to ensure it's populated as expected
print(df2.head())

# Saving DataFrame to a CSV file
df2.to_csv("yorub_sports_articles999.csv", index=False)
```
